# Añadiendo Medias Moviles (Simples y Exponenciales)

```python
# MM Normal
# Primero Crear las Medias Moviles
# Se crea una nueva columna de acuerdo a los periodos de las medias
# rolling es cuantos periodos va a correr hacia atras
# del rolling va a sacar el promedio
data2['MA20'] = data2['Close'].rolling(20).mean()
data2['MA150'] = data2['Close'].rolling(150).mean()

# MM Exponencial
# Parametro ewm (Exponential Moving Window)
# span es cuanto se va a expandir por periodos y se saca el promedio
data2['MA20-E'] = data2['Close'].ewm(span = 20).mean()
data2['MA150-E'] = data2['Close'].ewm(span = 150).mean()
```

```python
fig = go.Figure()

fig.add_trace(go.Candlestick(
                             x = data.index,
                             open = data2['Open'],
                             high = data2['High'],
                             low = data2['Low'],
                             close = data2['Close']
                             )
              )

# Añadimos la Media Movil de 20 periodos
fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data2['MA20'],
                         line = dict(color = 'blue',
                                     width = 1)
                         )
              )

# Añadimos la Media Movil de 150 Periodos
fig.add_trace(go.Scatter(
                         x = data2.index,
                         y = data2['MA150'],
                         line = dict(color = 'cyan',
                                     width = 1)
                         )
              )

fig.update_layout(xaxis_rangeslider_visible = False)

fig.show()
```

# Añadiendo Señales de Compra/Venta
*(Basadas en las Medias Moviles)*

```python
import numpy as np
```

```python
# Creamos una Nueva Columna
data2['Senal'] = 0.0
# El metodo where de la libreria numpy es para filtrar valores
# Va a estar filtrando valores donde la columna MA20 sea mayor a valores
# de la columna MA150, en caso de que cumpla la condicion, colocará el valor 1
# de lo contrario colocara el valor 0
data2['Senal'] = np.where(data2['MA20'] > data2['MA150'], 1.0, 0.0)
```

```python
data2
```

```python
# Crear la columna Posicion
# Queremos la señal especifica de solamente cuando se hizo el cruce
# Posicion estará ocupa con valores de 1 y -1
# Siendo 1 cuando es Compra y -1 cuando es Venta
# El metodo diff encuentra la diferencia entre valores, en este caso, sera con los de la columna
# de la columna senal
data2['Posicion'] = data2['Senal'].diff()
```

```python
data2
```

```python
# Por ahora sabemos que en este rango hay cambios
data2.loc['2016-12-05':'2016-12-23']
```

```python
fig = go.Figure()

fig.add_trace(go.Candlestick(
                             x = data2.index,
                             open = data2['Open'],
                             high = data2['High'],
                             low = data2['Low'],
                             close = data2['Close']
                             )
              )

# Ploteando Medias Moviles (Normales)
fig.add_trace(go.Scatter(
                         x = data2.index,
                         y = data2['MA20'],
                         line = dict(color = 'orange', width = 1)
                         )
              )
fig.add_trace(go.Scatter(
                         x = data2.index,
                         y = data2['MA150'],
                         line = dict(color = 'blue', width = 1)))

# Ploteando Señales de compra
fig.add_trace(go.Scatter(
                         x = data2[data2['Posicion'] == 1].index, # Solo ploteamos cuando aparezaca el valor 1 en la columna posicion
                         y = data2['MA20'][data2['Posicion'] == 1], # Ocupa de base la Media Movil y un filtro (columna Posicion) - Cuando es igual a 1 Marca la Compra
                         mode = 'markers', # Modo de Marcadores: markers
                         name = 'Compra', # Los marcadores se llamaran compra
                         marker = dict(size = 15, # Parametros del marcador: Tamaño del marcador
                                       color = 'black', # Color del marcador
                                       symbol = 'star-triangle-up' # Simbolo: Flecha o triangulo hacia arriba
                                       )
                         )
              )

# Plotenado Señales de Venta
fig.add_trace(go.Scatter(
                         x = data2[data2['Posicion'] == -1].index,
                         y = data2['MA20'][data2['Posicion'] == -1],
                         mode = 'markers',
                         name = 'Venta',
                         marker = dict(size = 15,
                                       color = 'cyan',
                                       symbol = 'star-triangle-down'
                                       )
                         )
              )

fig.update_layout(xaxis_rangeslider_visible = False)
fig.show()
```
