Como vemos, hemos descargado todas las columnas para dichos los activos, pero solo queremos las columnas 'High' de cada activo. Hagamos precisamente eso:

```python
# Fuente remota - Yahoo
assets = ['EURUSD=X', 'AUDEUR=X', 'USDGBP=X']

# Iteramos para obtener solo la columna 'High' de cada activo
data = pd.DataFrame()
for i in assets:
  data[i] = yf.download(i, interval='1d', start='2022-01-01', end='2022-02-02')['High']
```

# YFinance para Varios Tickers


```python
assets = ['EURUSD=X', 'EURGBP=X', 'AUDUSD=X']
```

```python
# Creamos un DataFrame vacio para contener los precios de cierre de
# todos los tickers descritos arriba
precio_cierre = pd.DataFrame()
```

```python
# Por cada activo o ticker seleccionado, va a descargar y almacenar el precio
# de cierre (Adj Close) de cada ticker
for ticker in assets:
  precio_cierre[ticker] = yf.download(ticker, period='14d', interval='1d')['Close']
```

```python
precio_cierre
```

```python
precio_cierre['EURUSD=X'].plot(figsize=(15,6))
```

Ahora si queremos almacenar todos las columnas de cada ticker (Open, High, Close, Adj Close, Volume, etc), tenemos que crear un DICCIONARIO. Sabemos ya que los diccionarios almacenan valores en "pares". En este caso la Clave o key sera el Ticker, y su valor sera el DataFrame (columnas de cada ticker).

```python
# Diccionario vacio
pares_data = {}
```

```python
for tickers in assets:
  pares_data[tickers] = yf.download(tickers, period='2y', interval='1d')
```

```python
pares_data
```
