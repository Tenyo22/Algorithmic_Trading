```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

```python
asset = 'BTC-USD'
start_date = '2015-01-01'
end_date = '2022-02-11'
data = yf.download(asset, interval='1d', start = start_date, end = end_date)
```

# Medias Moviles Simples

```python
# Instalar paquete Trend - Importamos indicador SMA
from ta.trend import SMAIndicator
```

```python
# Creamos el objeto derivado del indicador principal
# Como parametro requiere datos 'Close', los periodos a tomar y borrar o no las filas NaN
sma4 = SMAIndicator(data['Close'], 4, False)
sma12 = SMAIndicator(data['Close'], 12, False)

# Creamos una columna donde se populara con los valores de la Media Movil
data['SMA-4'] = sma4.sma_indicator() # Deriva el metodo para calcular
data['SMA-12'] = sma12.sma_indicator()
```

```python
data
```

```python
# Ploteamos Medias Moviles Simples

fig = go.Figure()

# Plot para el Precio 'Close'
fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data['Close'],
                         line = dict(color = 'grey'),
                         name = 'Precio de Cierre'
                         )
              )

# Plot para el SMA-4
fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data['SMA-4'],
                         line = dict(color = 'cyan'),
                         name = 'SMA-4'
                         )
              )

# Plot para el SMA-12
fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data['SMA-12'],
                         line = dict(color = 'blue'),
                         name = 'SMA-12'
                         )
              )
fig.show()
```

# Medias Moviles Exponenciales

```python
# Importar indicador EMAIndicator
from ta.trend import EMAIndicator
```

```python
data['EMA-20'] = EMAIndicator(data['Close'], 20, False).ema_indicator()
data['EMA-150'] = EMAIndicator(data['Close'], 150, False).ema_indicator()
```

```python
data
```

```python
# Ploteamos Medias Moviles Exponenciales

fig = go.Figure()

fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data['Close'],
                         line = dict(color = 'grey'),
                         name = 'Precio de Cierre'
                         )
              )

fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data['EMA-20'],
                         line = dict(color = 'cyan'),
                         name = 'EMA-20'
                         )
              )

fig.add_trace(go.Scatter(
                         x = data.index,
                         y = data['EMA-150'],
                         line = dict(color = 'blue'),
                         name = 'EMA-150'
                         )
              )

fig.show()
```
